```python
import pandas as pd

import numpy as np
# np.set_printoptions(threshold='nan')

%matplotlib inline
import matplotlib.pyplot as plt
```

```python
product_coords = np.loadtxt('data_for_demo_v2/product_plane_coords.txt', delimiter=' ')
```

```python
df = pd.read_csv('data_for_demo_v2/output_without_bbx.txt')
product_cols = ['product {}'.format(i) for i in range(len(product_coords))]
distances = df[product_cols].values # num_frames x num_products
```

```python
df[['frame_number'] + product_cols].head()
```

```python
FPS = 24
```

```python
df['seconds'] = df['frame_number'].apply(lambda x: x / FPS)
df.drop(['frame_number', 'Unnamed: 0'], axis='columns')
```

```python
distances -= np.min(distances, axis=1)[:, np.newaxis]
```

```python
MARKER_SIZE=5

fig = plt.figure(figsize=(16, 8))
ax1 = fig.add_subplot(111)

x = df.T.loc['frame_number']
for i in range(len(product_coords)):
    ax1.scatter(x, distances[:, i], s=MARKER_SIZE)

# plt.ylim(-1000, 2000)
plt.xlabel('frame')
plt.ylabel('distance from product')
plt.legend(product_cols)
```

```python
def plot_predictions(predictions, seconds=False):
    plt.figure(figsize=(16, 8))
    plt.ylabel('predicted product')
    if seconds:
        plt.xlabel('second')
        plt.plot([(i / FPS) for i in range(len(predictions))], predictions)
    else:
        plt.xlabel('frame')
        plt.plot(range(len(predictions)), predictions)

plot_predictions(df['predictions'])
```

```python
from scipy.signal import medfilt

smoothed_predictions = medfilt(df['predictions'], kernel_size=17)
plot_predictions(smoothed_predictions)
```

```python
np.save('data_for_demo_v2/smooth_prediction.npy', smoothed_predictions)
```
